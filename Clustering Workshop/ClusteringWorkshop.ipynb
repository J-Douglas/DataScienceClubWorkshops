In this example, we are using **k-means clustering** to classify flowers in the Iris dataset based on petal length and width. There are three types of flowers: Iris setosa, Iris versicolour, and Iris virginica. Learn more about the dataset here: https://archive.ics.uci.edu/ml/datasets/iris


```
# Importing libraries
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import datasets
from sklearn.cluster import KMeans
```

### Loading the dataset

The Iris dataset has sepal lengths, sepal widths, petal lengths, petal widths, and labels for each 150 Iris flowers. While this dataset is labelled, we will only be using the petal lengths and petal widths to create the clusters. We will compare our clusters to the actual labels of each flower.

```
dataset = datasets.load_iris() # loading the dataset
```

```
X = dataset.data[:, 2:4]  # petal lengths and petal widths
y = dataset.target # labels we will use for verifying our predictions
```

### Plotting the unlaballed data


```
plt.scatter(X[:,0], X[:,1])
plt.xlabel('Petal Length', fontsize=18)
plt.ylabel('Petal Width', fontsize=18)
plt.title('Unlabelled data', fontsize=30)
```

### Applying k-means clustering to dataset


```
model = KMeans(n_clusters = 3, random_state=21) # Setting the number of clusters to 3. Note: random_state is to make the model predictions repeatable
model.fit(X) # Running the k-means algorithm
```

```
# Clustering predictions
predictions = model.labels_ 

# Ploting the identified clusters and comparing with the answers
fig, axes = plt.subplots(1, 2, figsize=(16,8))

# Plotting actual classes
axes[0].scatter(X[:, 0], X[:, 1], c=y, cmap='Set1', edgecolor='k', s=150)
axes[0].set_xlabel('Petal length', fontsize=18)
axes[0].set_ylabel('Petal width', fontsize=18)
axes[0].tick_params(direction='in', length=10, width=5, colors='k', labelsize=20)
axes[0].set_title('Actual', fontsize=30)

# Plotting predictions
axes[1].scatter(X[:, 0], X[:, 1], c=predictions, cmap='Set3', edgecolor='k', s=150)
axes[1].set_xlabel('Petal length', fontsize=18)
axes[1].set_ylabel('Petal width', fontsize=18)
axes[1].tick_params(direction='in', length=10, width=5, colors='k', labelsize=20)
axes[1].set_title('Predicted', fontsize=30)
```
