## Pandas

Pandas is a data manipulation and analysis library. It is most commonly used for handling and examining datasets. The two main data structures that Pandas has are Series and Dataframes. A Series is an array that can hold integers, strings, floats, Python objects, and more. A Dataframe is a 2-dimensional labelled structure that contains columns, which can be of different types. You can think of Dataframes as spreadsheets or tables.

You can follow this great tutorial which has more thorough documentation [here](https://pandas.pydata.org/pandas-docs/stable/user_guide/10min.html).

```python
import pandas as pd
```

```python
# How to create a Series
S = pd.Series([4,7,8,12]) 
print(S)
```

```python
# How to create a Dataframe
D = pd.DataFrame({ 
        "A": 1.0,
        "B": pd.Timestamp("20211119"),
        "C": pd.Categorical(["test", "train", "test", "train"]),
        "D": "foo",
})
print(D)
```

## NumPy

NumPy is a Python library that is used to perform highly optimized operations on arrays and matrices. The core object in NumPy library is the ndarray.

This is a comprehensive yet concise resource about NumPy that I recommend you check out: https://colab.research.google.com/github/jakevdp/PythonDataScienceHandbook/blob/master/notebooks/02.00-Introduction-to-NumPy.ipynb#scrollTo=0lf7InVRocUh. 

```python
import numpy as np
```

```python
# Makes an array of 0's
np.zeros(4, dtype=int)
```

```python
# Makes an array of 1's
np.ones(4, dtype=int)
```

```python
# Makes an evenly spaced array using a start, end, and number of samples
np.linspace(0, 1, 5)
```

```python
# Makes an evenly spaced array using a start, end, and step size
np.arange(3,12,2)
```

```python
# Making an identity matrix
np.eye(3)
```

```python
# Making a NumPy array
x = np.array([1, 2, 3])
print(x)
print(x.shape)
```

```python
# Reshaping array
x = x.reshape((1, 3))
print(x)
print(x.shape)
```

```python
# Concatenating arrays
y = np.array([1, 2, 3])
z = np.array([3, 2, 1])
w = np.concatenate([y, z])
print(w)
print(w.shape)
```

```python
# Transpose arrays
a = np.array([[0,4,6],[8,np.NaN,12]]) # Note: NaN stands for Not a Number and essentially means that the value is "undefined"
print(a)
print(np.transpose(a))
```

## Matplotlib/Seaborn

Matplotlib is a Python plotting library and extends NumPy. It provides a comprehensive API for various static and animated visualizations. Some types of plots that you can create include scatter plots, box plots, line graphs, pie graphs, and more. Seaborn is another statistical graphing library that is built upon Matplotlib. It can aid your exploration and understanding of data.

You can learn more about the Matplotlib library documentation [here](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.html). You can learn more about the Seaborn library documentation [here](https://seaborn.pydata.org/introduction.html).

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Plotting age vs. cholestorol data with Matplotlib
plt.figure()
plt.scatter(df['age'],df['chol']) # Scatter plot function
plt.grid() # Provides grid lines
plt.xlabel("Age in yrs") # X-title
plt.ylabel("Cholesterol in mg/dL") # Y-title
plt.title("Age vs Cholesterol") # Figure title
```

```python
# Useful seaborn tool to plot several features against each other
sns.set(style="ticks")
sns.pairplot(df, x_vars=['age','cp','trestbps','chol','fbs'], y_vars=['age','cp','trestbps','chol','fbs'], palette = ('b', 'r'), hue="target", height=2.5)
plt.show()
```

## TensorFlow/Keras

TensorFlow is an open-source machine learning library that is used for creating artificial neural networks (ANNs). Artificial neural networks are a type of machine learning which are composed of layers of neurons and are used to solve a wide range of problems including classifcation tasks and sequence recognition. They're structure is inspired by biological neural networks and the process of how animals learn. You can learn more about TensorFlow [here](https://www.tensorflow.org/learn). 

![Neural Network](https://drive.google.com/uc?export=view&id=1GIbRwPD49jxptyAerjFlmtsNmuzoopx7)

Keras is an interface for TensorFlow and it contains implementations of several common building blocks in neural networks including layers, activation functions, optimizers, and more.

If you want to learn more about theory and application of neural networks, we have several videos on this topic such as [Neural Networks and Computer Vision](https://www.youtube.com/watch?v=Y_YsKhBooOc&ab_channel=UWaterlooDataScience) and [Introduction to Neural Networks in TensorFlow 2.0](https://www.youtube.com/watch?v=a_RkE897-Kw&ab_channel=UWaterlooDataScience).

Note that PyTorch is another alternative machine learning library that accomplishes many of the same things as TensorFlow, so that is another library you might want to check out.

### TensorFlow ANN Basics

```python
import tensorflow as tf
from tensorflow import keras
```

```python
from keras.models import Sequential # The Sequential object is used to create a neural network with a linear stack of layers
from keras.layers import * # Layers module
```

```python
# Creating a simple sequential model
model = tf.keras.Sequential()
model.add(tf.keras.Input(shape=(3,))) 
model.add(tf.keras.layers.Dense(3, activation='relu'))
model.add(tf.keras.layers.Dense(4, activation='relu'))
model.add(tf.keras.layers.Dense(2, activation='softmax'))
```

You can then compile your model with the "compile" function. You will need to specify the optimizer and loss functions. If you want to learn more about what this means, check out the workshops mentioned before!

To train our model, we use the "fit" function and specify the training data, training labels, the batch size, and the number of epochs. To make predictions with our model, we use the "predict" function and specify the testing samples.
